```python
import dash
from dash import dcc, html, Input, Output
import plotly.express as px
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)

df = pd.DataFrame({
    "Year": np.repeat(np.arange(2010, 2020), 20),
    "Sales": np.random.randint(200, 1000, 200),
    "Recession": np.random.choice([0,1], 200),
})
```

```python
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Automobile Sales Dashboard")
])
```

```python
app.layout = html.Div([
    html.H1("Automobile Sales Dashboard"),

    html.H2("Select Report Type"),

    dcc.Dropdown(
        id="report-type",
        options=[
            {"label": "Recession Report", "value": "recession"},
            {"label": "Yearly Report", "value": "yearly"}
        ],
        value="recession"
    )
])
```

```python
app.layout = html.Div([
    html.H1("Automobile Sales Dashboard"),

    dcc.Dropdown(
        id="report-type",
        options=[
            {"label": "Recession Report", "value": "recession"},
            {"label": "Yearly Report", "value": "yearly"}
        ],
        value="recession"
    ),

    html.Div(id="output-container", className="chart-container")
])
```

```python
@app.callback(
    Output("output-container", "children"),
    Input("report-type", "value")
)
def update_graph(selected):

    # 📌 Q5: Recession report graph
    if selected == "recession":
        filtered = df[df["Recession"] == 1]
        fig = px.line(filtered, x="Year", y="Sales",
                      title="Recession Report - Sales Trend")

    # 📌 Q6: Yearly report graph
    else:
        fig = px.line(df, x="Year", y="Sales",
                      title="Yearly Report - Sales Trend")

    return dcc.Graph(figure=fig)
```
